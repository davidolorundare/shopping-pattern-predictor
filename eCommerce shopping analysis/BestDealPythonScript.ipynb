### Data Cleaning Tasks that will be done.

From our observations of the dataset:
- We note it has 10000 rows and 34 columns, some of which
we observe have missing data (.i.e. NaN).

- Alot of these columns are of float datatype, but we note 
that every product is suppose to have a value of 1; which 
indicates they WERE bought, or a value of 0; which indicates
they were NOT bought. This implies that columns with float 
(datatype columns) SHOULD be INTEGER. HENCE, ANY column that 
is NOT an integer (e.g. float, object) SHOULD BE converted to integer.

- Any product values other than these 1 or 0 (e.g. 10,11, 19) SHOULD have their rows removed.

- We also note that the CustomerAge column ranges from 0 to 923,
but we were told earlier that the Ages should are SUPPOSE to
range from 18 to 150. HENCE, we will slice the data and ONLY MAKE
USE OF (or work with) data within the range of CustomerAge 18 TO 150.

```python
# Based on the above four notes, we begin cleaning the data:
```

- First, we remove NaN data.

```python
# We drop any missing data (.i.e. NaN)
cleandata = dirtydata4bestdeal.dropna()
cleandata.shape
```

```python
dropped_cols = 10000 - 9432
print("We observe: {} rows with NaN data were removed.".format(dropped_cols))
```

```python
cleandata.head()
```

- Next, we convert all the columns datatypes to integers.

Ideally we'd like to iterate through the dataset and
convert each column to integer, but four columns: 'SonyTV42LED',
'XBOX360', 'DellLaptop', and 'BoseSoundSystem' have String 
as their datatype; even though their contents are all numbers.

We need to convert those object (string) columns dataype first 
to numeric.
Attempting to use the usual Series astype() function directly
results in a "ValueError: could not convert string to float:" 
message.

Hence, we use a different approach, we use Pandas' to_numeric()
function to coerce each of these four column into numeric values, 
any of the column values that cannot be coerced (e.g. if the value 
is an empty string) is converted into a NaN value.

```python
# First we store the names of the object columns we want to coerce to numeric
columns_to_coerce = ['SonyTV42LED','XBOX360', 'DellLaptop', 'BoseSoundSystem']

# Next, we iterate through each column; coercing its values to numeric
for column in columns_to_coerce:
    cleandata[column] = cleandata[column].apply(pd.to_numeric, errors='coerce')
```

We check the datatype of the dataset, noting that those four columns
( 'SonyTV42LED','XBOX360', 'DellLaptop', and 'BoseSoundSystem') have 
had their datatype now converted from object to float64

```python
cleandata.dtypes
```

Before we continue, from the previous coercion operation, some column 
values would have been assigned NaN, because either their values were 
empty-strings or another non-numeric value.

```python
# Hence, we drop the rows associate with those values
cleanedData = cleandata.dropna()

# This gives us a result of:
cleanedData.shape
```

Having converted those object columns to numeric and stripped
any NaN data from it, we can now convert ALL columns into integers 
(to strip the decimal notation on some columns)

```python
# To do this, we first use python list-comprehension to get all the columns to iterate over
datacolumns = [column for column in cleanedData]

# Finally we convert all the columns to integer and store the result
# back in our Dataframe
cleanedData[datacolumns] = cleanedData[datacolumns].astype(int)
```

```python
# We inspect the converted data, noting
# all its column datatypes are now integer.
cleanedData.dtypes
```

```python
# Taking alook at the cleaned data itself
cleanedData.head()
```

- Removal of the rows of any product-column whose value is not 1 or 0.

```python
# First, let us take note of the number of rows in the data
# we have cleaned so far
cleanedData.shape
```

We'd like to iterate through each product-column and for any
value we see that is not a 1 or 0 (i.e. any value greater 
than 1; like 10,11, 19, etc) their corresponding row should be removed.

```python
# Let us first get all the product-columns in the data
productColumns = cleanedData.iloc[:,2:]
productColumns.head(3)
```

```python
# Then let us get all the names of those product columns
productColumnNames = [column for column in productColumns]
```

Next, we iterate through each of these product
columns and use a booelan conditional check to 
identify product columns with values greater than 1.

```python
# If the value is greater than 1 (i.e. it is not the 
# desired 0 or 1 value) then we drop the entire row
# corresponding to that value.
# We do this 'inplace' making the removals to be done 
# within the exact same cleanedData Dataframe on each iteration.

for product in productColumnNames:
    cleanedData.drop(cleanedData[cleanedData[product] > 1 ].index, inplace=True)
```

```python
# We examined the resulting cleaned Dataframe
cleanedData.head()
```

```python
# Taking note of the number of rows that were removed with this cleaning
cleanedData.shape
```

```python
dropped_columns = 9430 - 9204
print('This implies that {} rows were dropped for product-columns whose value was not 1 or 0'.format(dropped_columns))
```

- Slice dataset to extract and use only data having CustomerAge value ranges between from 18 and 150.

Our Final data-cleaning task is to use only part of the data that has CustomerAge range
between 18 and 150, given that any other values outside of these range indicate invalid 
data. We thus slice our earlier 'cleanedData' to take a subset where CustomerAge is 18 to 150.

```python
# To do this, first we create the boolean condition to subselect
# data wherein the CustomerAge is between 18 and 150
customerSegment = (cleanedData['CustomerAge'] >= 18) & (cleanedData['CustomerAge'] <= 150)

# Then we slice the dataset to extract only data in that CustomerAge range (18 - 150)
cleanWorkingData = cleanedData[customerSegment]

# The resulting data is what we will now be working with henceforth for the later questions of the project, thus:
cleanWorkingData.head()
```

### Requirement #3.1
Create a model to predict if customer X will  buy the newest Sony TV 60 LED

```python
# Add your code and commentary here for Requirement #3.1

# First we import all the machine learning modules we will need for the task.
from sklearn.cross_validation import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import preprocessing

# To make the predictions, we use idenifiable information about the customer 
# in order to create the predictive model - We are performing a classification-task
# : The Customer's Zipcode and Age are used predict whether a new 
# customer will buy the SonyTV60LED - 1 , or not buy - 0
 
# X will contain our feature variables; 'Zipcode' and 'CustomerAge'
X = cleanWorkingData.values[:, 0:2].astype(float)

# Y will contain our target variable; 'SonyTV60LED'
Y = cleanWorkingData.values[:, 16]    

# Due to the non-uniformity and high dimensionality of the column-values features
# Zipcode (5-digits) and CustomerAge(2-digit) versus the SonyTV60LED column-values
# (1-digit) target, the values in these first two columns must be scaled.

# The numerical difference between the scaled and non-scaled
# versions of the data, is that the non-scaled data has non-unit 
# standard deviation, while the scale data has unit standard deviation
# and zero mean; the performance of several machine learning algorithms
# depend on whether the data is uniform and scaled.

# Hence, we scale our training dataset features
X_scaled = preprocessing.scale(X)

# Our classification task will require a Supervised Machine Learning approach
# We make use of a DecisionTree Classifier, and build its model.

# As good practice, we split our dataset into training and test dataset for optimal use of the data
# Making the training data be 70% of the whole data and the remaining 30% be the testing data
X_train, X_test, y_train, y_test = train_test_split( X_scaled, Y, test_size = 0.3, random_state = 100)

# Finally, we build the DecisionTree Classifier model, using the Gini Criterion
classifier_gini = DecisionTreeClassifier(criterion = "gini", random_state = 100, max_depth=3, min_samples_leaf=5)
# and train the classification model for use.
classifier_gini.fit(X_train, y_train)
```

```python
# Add your code and commentary here for Requirement #3.2

# First we import all the machine learning modules we will need for the task.
%matplotlib inline
import matplotlib.pyplot as plotter
from sklearn import preprocessing
from sklearn.cluster import KMeans


# To make the segmentation, we use an Unsupervised Machine Learning approach
# to cluster the data into groups based on some of its features
# The KMeans Clustering machine learning algorithm is use to perform this task.

# We decide to use only two features to perform the segmentation - 'ZipCode'
# and 'CustomerAge'

# We keep all the features in a 
# two-dimensional matrix for easier computation
feature1 = cleanWorkingData['ZipCode'].values
feature2 = cleanWorkingData['CustomerAge'].values

data_segment_matrix = np.matrix(zip(feature1,feature2))

# As explained in the previous Requirement, we need to scale the data
# to resolve non-uniformity and high-dimensionality problems.
segment_data_scaled = preprocessing.scale(data_segment_matrix)

# Finally we use the KMeans algorithm to cluster the data into groups,
# segmenting into 2 groups.
kmeansCluster = KMeans(n_clusters=2).fit_predict(segment_data_scaled)

#Plotting the two groups of clusters on a graph
plotter.scatter(segment_data_scaled[:, 0], segment_data_scaled[:, 1], c=kmeansCluster, s=70, cmap='viridis');
plotter.title("-Segmentation based on (scaled) Zip Code and Customer Age data-")
#plotter.xlabel('Zip Code')
#plotter.ylabel('Customer Age')
plotter.show()

# Alternative plotting with Axes labels
# fig = plotter.figure()
# fig.suptitle('-Segmentation of (scaled) Customer data into Groups-', fontsize=14, fontweight='bold')

# ax = fig.add_subplot(111)
# fig.subplots_adjust(top=0.85)
# ax.set_title('Zip Code and Customer Age Segmentation')

# ax.set_xlabel('Customer Age')
# ax.set_ylabel('Zip Code')
             
# ax.plot(segment_data_scaled[:, 0], segment_data_scaled[:, 1], 'o')
# plotter.show()
```
